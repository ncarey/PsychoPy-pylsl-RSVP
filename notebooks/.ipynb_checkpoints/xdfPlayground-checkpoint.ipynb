```python
import sys
import os.path
from matplotlib import pyplot as plt

project_path = '/home/idies/workspace/Storage/ncarey/persistent/PULSD/PsychoPy-pylsl-RSVP/'
sys.path.append(project_path)

from xdfReadLib.xdf import load_xdf


```

```python
xdf_file_path = os.path.join(project_path, 'recordings/CheckerVEP_4hz.xdf')

xdf = load_xdf(xdf_file_path, verbose=False)

print(xdf[0][0].keys())

#I think time series is the frame count, time stamps is the actual clock
time_series = xdf[0][0]['time_series']

time_stamps = xdf[0][0]['time_stamps']
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.scatter(time_series, time_series, marker='o', s=6)

ax1.set_xlim(0,500)
ax1.set_ylim(0,500)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)

time_stamps_plot = time_stamps - 4840

ax1.scatter(time_stamps_plot, time_stamps_plot, marker='o', s=6)

ax1.set_xlim(40,42)
ax1.set_ylim(40,42)
```
